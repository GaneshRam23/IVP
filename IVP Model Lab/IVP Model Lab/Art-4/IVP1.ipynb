```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = cv2.imread(r'C:\Users\SNUC\Downloads\Art.jpg')
```

```python
global_threshold = np.vectorize(lambda x: 0 if x<127 else 255)(img)
plt.imshow(global_threshold)
```

```python
import cv2

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


adaptive_thresholded_image = cv2.adaptiveThreshold(
    gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blockSize=11, C=2
)

plt.imshow(adaptive_thresholded_image)
```
