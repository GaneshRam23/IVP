```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = cv2.imread(r'C:\Users\SNUC\Downloads\Art.jpg')
```

```python
img.shape
```

```python
pixels = np.reshape(img, (-1, 3))
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(pixels)
```

```python
wcss = []
cls = []
for i in range(1,10):
    cls.append(KMeans(n_clusters=i))
    cls[i-1].fit(pixels)
    wcss.append(cls[i-1].inertia_)
```

```python
plt.plot(range(1,10),wcss)
plt.scatter(range(1,10),wcss,c='red')
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(pixels)
```

```python
cluster_labels = []
cluster_centers = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(pixels)
    wcss.append(kmeans.inertia_)
    cluster_labels.append(kmeans.labels_)
    cluster_centers.append(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(8, 6))
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')

for i in range(9):
    plt.figure(figsize=(8, 6))
    clustered_pixels = cluster_centers[i][cluster_labels[i]]
    clustered_image = np.reshape(clustered_pixels, img.shape).astype(np.uint8)
    plt.imshow(clustered_image)
    plt.title('Clustered Image (K = {})'.format(i+1))
    plt.axis('off')
    plt.show()
```
