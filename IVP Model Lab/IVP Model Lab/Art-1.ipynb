```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path='/home/snucse/Downloads/Art.jpg'
image=cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image[:,::-1]),plt.title("Original Image")
plt.show()
```

# Global Thresholding

```python
global_threshold_value = 127 
_, global_thresholded_image = cv2.threshold(image, global_threshold_value, 255, cv2.THRESH_BINARY)
```

# Adaptive Thresholding

```python
adaptive_thresholded_image = cv2.adaptiveThreshold(
    image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blockSize=11, C=2
)
```

```python
##Displaying original images and threshold images
plt.figure(figsize=(12, 6))

plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(global_thresholded_image, cmap='gray')
plt.title('Global Thresholding')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(adaptive_thresholded_image, cmap='gray')
plt.title('Adaptive Thresholding')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/home/snucse/Downloads/Art.jpg' )
```

# Converting BGR to RGB

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
pixels = image.reshape((-1, 3)).astype(np.float32)
```

```python
num_clusters = 5
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
_, labels, centers = cv2.kmeans(pixels, num_clusters, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
centers = np.uint8(centers)
```

```python
segmented_image = centers[labels.flatten()]
segmented_image = segmented_image.reshape(image.shape)
```

# Displaying original and segmented image

```python
plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(segmented_image)
plt.title('Segmented Image ({} clusters)'.format(num_clusters))
plt.axis('off')

plt.tight_layout()
plt.show()
```
